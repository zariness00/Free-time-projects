```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
```

How effective is NaiveBayes at telling the difference between a baseball email and a hockey email❓

```python
emails = fetch_20newsgroups(categories = ["rec.sport.baseball", "rec.sport.hockey"])
```

```python
#printing different categories
emails.target_names
```

```python
#what kind of email is there?
emails.data[5]
```

```python
#identifying the label
emails.target[5]
#the label is for hockey
```

```python
train_emails = fetch_20newsgroups(categories = ["rec.sport.baseball", "rec.sport.hockey"], subset = "train", shuffle = True, random_state = 108)
```

```python
test_emails = fetch_20newsgroups(categories = ["rec.sport.baseball", "rec.sport.hockey"], subset = "test", shuffle = True, random_state = 108)
```

```python
#transforming into a list of counts
counter = CountVectorizer()
counter.fit(test_emails.data + train_emails.data)
```

```python
train_counts = counter.transform(train_emails.data)
test_counts = counter.transform(test_emails.data)
```

```python
classifier = MultinomialNB()
classifier.fit(train_counts, train_emails.target)
```

```python
#classify all the emails in the test set and their labels
classifier.score(test_counts, test_emails.target)
```
